maxpool只有最大的地方有梯度，avgpool梯度均分

```python
import torch
import torch.nn as nn

avgpool2d = nn.AvgPool2d(2, stride=2)
maxpool2d = nn.MaxPool2d(2, stride=2)
x = torch.rand(1, 3, 4, 4, requires_grad=True)
print(x)
y = maxpool2d(x)
print(y)
loss = (y**2).sum()
loss.backward()
print(x.grad)
x.grad.zero_()
y = avgpool2d(x)
loss = (y**2).sum()
loss.backward()
print(x.grad)
```
